```python
## Data Science Masters Session 10 Assignment Q2.2 Scatterplot Fare Age gender
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)

titanic = titanic.dropna(subset=['sex'])

mapping = {'male' : 'blue', 'female' : 'red'}
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5, marker ='x', c=titanic['sex'].map(mapping))

plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Fare Age Scatterplot by gender Blue is Male and Red is Female")

plt.show()
```
