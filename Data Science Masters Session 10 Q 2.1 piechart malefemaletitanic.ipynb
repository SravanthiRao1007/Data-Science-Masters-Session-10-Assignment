```python
## Data Science Masters Session 10 Assignment Q2.1 Pie graph of male female proportion 
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)

exp_labels=['Female','Male']
plt.axis('equal')
plt.pie(titanic.pclass.groupby(titanic.sex).sum(),labels=exp_labels,shadow=True, autopct='%1.1f%%',radius=1.5)
print("\n The male female proportion in the given data is as below ")
plt.show()
```
